```python
import pandas as pd
import ast

df = pd.read_csv(r"C:\Users\KIIT0001\Desktop\Projects\Tweets.csv")
# Basic info
print("Shape (rows, columns):", df.shape)
print("\nColumns:\n", df.columns)
```

```python
missing = df.isna().sum().sort_values(ascending=False)
missing
```

```python
drop_cols = [
    "tweet_id_str",
    "airline_sentiment_gold",
    "negativereason_gold",
    "user_timezone"]
df = df.drop(columns=drop_cols, errors="ignore")
```

```python
df = df.dropna(subset=["text"])
```

```python
df = df.dropna(subset=["tweet_coord"])
```

```python
df["tweet_coord"].apply(type).value_counts()
```

```python
##convert strings to list
df["tweet_coord"] = df["tweet_coord"].apply(ast.literal_eval)
```

```python
df[["lat", "lon"]] = pd.DataFrame(df["tweet_coord"].tolist(), index=df.index)
##converted tweets_coord[-10,12] to latitute(lat) and longitude(lon) separate for better use in streamlit
```

```python
print("Final shape:", df.shape)
df[["lat", "lon"]].isna().sum()
```

```python
df.to_csv(r"C:\Users\KIIT0001\Desktop\Projects\Tweets_cleaned.csv", index=False)
```
